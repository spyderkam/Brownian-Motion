# <center> Analysis of Trapping Length </center>

$$ l_{\mathrm{p}} \simeq 2\times\texttt{per}^2 \tag{1}\label{1}$$

where $l_{\mathrm{p}}$ is the persistence length and $\texttt{per}$ is a computational variable. The approximation is accurate for larger $\texttt{per}$.

````python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
import scipy.io as sio
from IPython.display import display_pdf
````

***
# Parameters A

- $\texttt{per} = \{[0,2],\,2.5,\,3\}$ - Make this $0$ to have a freely jointed walk, otherwise the bigger this is the stiffer the chain is. 

- $\Delta{\texttt{per}} = 0.2$

- $\texttt{iter}=1,\!000$ - number of walks to simulate

- $\texttt{vis}=0$

````python
per = [0, 0.2, 0.4, 0.6, 0.8, 1, 1.2, 1.4, 1.6, 1.8, 2, 2.5, 3]
mean_traptrack = [102.0820, 102.4240, 101.5220, 100.7830, 100.8260, 96.8860, 110.0640, 125.8830, 145.4550, 170.3400, 210.6070, 301.5370, 419.9740]
````

### Plotting the mean trapping length as a function of $\mathrm{per}$

````python
def f(x,a,b,c):
    return a*(x)**b + c

plt.plot(per, mean_traptrack, 'b--', label='data')

popt, pcov = curve_fit(f, per, mean_traptrack)
plt.plot(per, f(per, *popt), 'r-', label='fit: a=%5.3f, b=%5.3f, c=%5.3f' % tuple(popt))

plt.legend()
plt.show()
````

````python
xrange = np.linspace(0,3)  # len(xrange) = 50 by default
pretty_curve = f(xrange, popt[0], popt[1], popt[2])

plt.scatter(per, mean_traptrack, color='blue', label='data')
plt.plot(xrange, pretty_curve, color='red', label=rf'fit: $a=%5.3f,~b=%5.3f,~c=%5.3f$' % tuple(popt))

plt.xlabel(r"per", fontsize=20)
plt.ylabel(r"$\left\langle \mathrm{trapping~length} \right\rangle = ax^b+c$", fontsize=16.5)

plt.tight_layout()
plt.legend()
#plt.savefig('mean_traplength_vs_per.png', bbox_inches='tight')
plt.show()
````

### Plotting the mean trapping length as a function of $2\times\mathrm{per}^2$

````python
tpt = 2*np.array(per)**2   #2per^2
plt.scatter(tpt, mean_traptrack, color='black')
plt.xlabel(r"$2\times\mathrm{per}^2$", fontsize=20)
plt.ylabel(r"$\left\langle\mathrm{trapping~length}\right\rangle$", fontsize=20)

plt.tight_layout()
plt.title("$\mathrm{iter} = 1,\!000$", fontsize=20)
#plt.savefig('plots/mean_traplength_vs_2per2_iter=1000.pdf', bbox_inches='tight')
plt.show()
````

### Plotting the mean trapping length as a function of $\left\langle{l_{\mathrm{t}}}\right\rangle  \cdot \left(2\times\mathrm{per}^2\right)^{-1}$

````python
ydata = np.array(mean_traptrack)/(2*np.array(per)**2)
plt.scatter(tpt, ydata)

plt.xlabel(r"per", fontsize=20)
plt.ylabel(r"$\left\langle l_{\mathrm{t}} \right\rangle \cdot \left(2\times\mathrm{per}^2\right)^{-1}$", fontsize=20)

plt.ylim(0,100)
plt.show()
````

<u>Klotz's instructions:</u> Divide the $x$ values of each one by the mean trapping length so instead of peaking at $120$ or whatever they just peak at $x = 1$ (or close to it), and then normalize them.

The $x$ values *were* `m1` and `m1/tpt`. *Now* they are `m1/mean_traptrack` and `(m1/tpt)/mean_traptrack`.

````python
ave_lt = np.array(mean_traptrack)

for i in range(len(per)):
    plt.plot(m1[i]/ave_lt[i], m2[i])
    #plt.title(f"per = {per[i]}", fontsize=20)
    plt.ylabel("# of simulations", fontsize=20)
    plt.xlabel(r"$l_{\mathrm{t}} / \left\langle l_{\mathrm{t}} \right\rangle$", fontsize=20)
    #plt.savefig(f'plots/histograms/iter=10000/all_in_one_sim_v_lt_per_ave_lt.pdf', bbox_inches='tight')
    plt.show()
````

````python
for i in range(len(per)):
    plt.plot((m1[i]/tpt[i])/ave_lt[i], m2[i])
    #plt.title(f"per = {per[i]}", fontsize=20)
    plt.ylabel("# of simulations", fontsize=20)
    plt.xlabel(r"$\left[l_{\mathrm{t}} \cdot \left(2\times\mathrm{per}^2\right)^{-1}\right] / \left\langle l_{\mathrm{t}} \right\rangle$", fontsize=20)
    #plt.savefig(f'plots/histograms/iter=10000/all_in_one_sim_v_lt_per_tpt_per_ave_lt.pdf', bbox_inches='tight')
plt.show()
````

Based on the plot above, as $\texttt{per}$ increases, the following decreases; 

$$
    \left[l_{\mathrm{t}} \cdot \left(2\times\texttt{per}^2\right)^{-1}\right] / \left\langle l_{\mathrm{t}} \right\rangle
    \tag{2}\label{2}
$$

as expected since $\texttt{per}$ is in the denomicator. In otherwords, as the persistance length increases, $\eqref{2}$ decreseases since,

$$
    \left[l_{\mathrm{t}} \cdot \left(2\times\texttt{per}^2\right)^{-1}\right] / \left\langle l_{\mathrm{t}} \right\rangle 
    \simeq \dfrac{l_{\mathrm{t}}}{l_{\mathrm{p}} \left\langle l_{\mathrm{t}} \right\rangle}
    \tag{3}\label{3}
$$
